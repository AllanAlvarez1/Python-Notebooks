```python
#First Step is to import matplotlib
import matplotlib
```

```python
#Most of the math plotting utilities lies under the pyplot submodule as plt
import matplotlib.pyplot as plt
```

```python
#import numpy to work with arrays in math
import numpy as np
```

```python
#Example 1
#Draw a line from the Origin to another position
xPlane = np.array([0, 6]) #X plane
yPlane = np.array([0, 250]) #Y plane

plt.plot(xPlane, yPlane)
plt.show()
```

The plot() function is used to draw points in a diagram by going from point to point.
Paramater 1 is the X-axis
Paramater 2 is the Y-axis

```python
#Example 2
#Draw another line
xPoints = np.array([1,8])
yPoints = np.array([4, 20])

plt.plot(xPoints, yPoints)
plt.show()
```

To plot only with the points showing using the shortcut string notation'o' which means rings

```python
xPoints = np.array([1,8])
yPoints = np.array([4, 20])

plt.plot(xPoints, yPoints, 'o')
plt.show()
```

To plot multiple points you simply need to have the same number of points in both axis

```python
#Example 3
#Multiple points
xPoints = np.array([1, 2, 6, 8])
yPoints = np.array([3, 8, 1, 10])

plt.plot(xPoints, yPoints)
plt.plot(xPoints, yPoints, 'o')
plt.show()
```

If the x points are not specified then by default they are the values:
0, 1, 2, 3, etc... depending on the length of y points given


```python
#Example 4
#Leave out the x points and plot the graph
yPoints = np.array([3, 8, 1, 10, 7, 9, 20])
plt.plot(yPoints)
plt.show()
```

You can use the keyword 'marker' to emphasize a point with a specific marker like o or x

```python
#Example 5
#Draw a line and mark the points along the line
import matplotlib.pyplot as plt
import numpy as np

yPoints = np.array([1,3,5,6,7,9])

plt.plot(yPoints, marker = 'o')
plt.show()
```

You can use the shortcut string notation parameter to specify the marker. This marker is called fmt and it is written with the syntax:
marker|line|color

```python
import matplotlib.pyplot as plt
import numpy as np

yPoints = np.array([1,3,5,7,9])

plt.plot(yPoints, 'o:g')
plt.show()
```

The line values can be - : -- and -.
The color values can be r g b c m y k w

To adjust the size of the marker you can use the markersize argyment or ms to set the size

```python
#Example 6
#Adjust the size of the marker
yPoints = np.array([1,3,5,7,9])

plt.plot(yPoints, 'o:c', ms = 20)
```

You can also use the argument 'markeredgecolor' or 'mec' to set the color of the edge of the markers

```python
#Example 7
#Set the edge color of the markers
import matplotlib.pyplot as plt
import numpy as np

yPoints = ([1,3,5,7,9])

plt.plot(yPoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

You can use the markerfacecolor or mfc argument to set the color of the edge inside the markers

```python
#Example 8
#Set the color inside the edge of the marker
import matplotlib.pyplot as plt
import numpy

yPoints = ([1,3,5,7,9])

plt.plot(yPoints, 'o-', ms = 20, mec = 'c', mfc = '#ff56ba')
plt.show()
```

You can also use hexadecimal color values to color your face or edges

You can use the argument 'linestyle' or ls to change the style of the plotted line

```python
#Example 8
#change the style of the plotted line
import matplotlib.pyplot as plt
import numpy as np

yPoints = ([1,3,5,7,9])

plt.plot(yPoints, 'o', ls = 'dashed', ms = 20, mfc = 'hotpink', mec = 'c')
plt.show()
```

You can use dotted, dashed, dashdot, solid, or None as your arguments for you linestyle values

The line color can be changed with the 'color' argument which can be shortened to c

```python
#Change the color of the line

import matplotlib.pyplot as plt
import numpy as np

yPoints = ([1,3,5,7,9])

plt.plot(yPoints, marker = 'o', linestyle = 'dashdot', color = 'green', ms = 20, mfc = 'hotpink', mec = 'cyan')
plt.show()
```

You can use the argument 'linewidth' or 'lw' to change the width of a line with a value that is a floating number

```python
#Change the width of the line
import matplotlib.pyplot as plt
import numpy as np

yPoints = ([1,3,5,7,9])

plt.plot(yPoints, marker = 'o', mfc = 'hotpink', mec = 'cyan', linewidth = 5.4, color = 'green')
plt.show()
```

To plot multiple lines you add more plot functions

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(y1, 'o-r', color = 'blue', ms = 20, mfc = 'hotpink')
plt.plot(y2, 'o-b', color = 'green', ms = 20, mfc = 'cyan')

plt.show()
```
